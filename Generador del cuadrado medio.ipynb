```python
from tabulate import tabulate

def middle_square_method(seed, digits=4, iterations=200):
    seen = set()  # Para evitar ciclos repetitivos
    random_numbers = []
    number = seed
    iterations_list = []  # Lista para almacenar el número de iteración
    num = []
    for i in range(iterations):
        squared = number ** 2  # Elevamos al cuadrado
        squared_str = str(squared).zfill(2 * digits)  # Aseguramos la longitud con ceros (longitud de 8 digitos)

        # Extraer los 4 dígitos centrales
        start = (len(squared_str) - digits) // 2
        middle_digits = int(squared_str[start:start + digits])

        # Si el número ya apareció antes se detener el proceso
        if middle_digits in seen:
            break

        seen.add(middle_digits)
        random_numbers.append(middle_digits)
        iterations_list.append(i + 1)  # Guardar el número de iteración
        number = middle_digits
        num.append(squared_str)

    # Normalización de los números generados
    if len(random_numbers) > 1:  # Evitar errores si solo hay un número
        normalized_numbers = (np.array(random_numbers) - min(random_numbers)) / (max(random_numbers) - min(random_numbers))
    else:
        normalized_numbers = np.array([0])

    # Calcular ciclo de vida
    ciclo_de_vida = len(random_numbers)

    # Crear DataFrame con la información generada
    df = pd.DataFrame({
        "x_j": iterations_list,
        "Semilla^2": num,
        "Número Generado": random_numbers,
        "Número Normalizado": normalized_numbers
    })

    # Creamos la tabla
    table = tabulate(df, headers="keys", tablefmt="grid", showindex=False)

    # Mostrar la tabla y el ciclo de vida
    print(table)
    print(f"\nCiclo de vida de la semilla {seed}: {ciclo_de_vida} iteraciones.")

    return df

seed = 8347 # semilla
df_result = middle_square_method(seed, digits=4, iterations=200)
```
