```python
from tabulate import tabulate
import numpy as np
import pandas as pd

def generador_congruencial(a, c, m, x0, n):
    numeros = []
    x = x0

    for _ in range(n):
        x = (a * x + c) % m # fórmula del generador congruencial
        numeros.append(x)

    normalizados = [xi / m for xi in numeros]

    return numeros, normalizados

def encontrar_periodo(numeros):
    for i in range(1, len(numeros)):
        if numeros[i] == numeros[0]:  # Encuentra la primera repetición
            return i
    return len(numeros)  # Si no hay repetición, retorna la longitud de la lista

# Parámetros del generador
a = 21 # Multiplicador
c = 15  # Incremento
m = 31  # Módulo
x0 = 21  # Semilla
n = 50  # Cantidad de números a generar

# Generación de números
numeros, normalizados = generador_congruencial(a, c, m, x0, n)

# Encontrar periodo del generador
periodo = encontrar_periodo(numeros)

# Crear tabla
tabla = list(zip(range(1, n+1), numeros, normalizados))
headers = ["i", "x_i", "x_{i+1}"]

print("Tabla de números generados:")
print(tabulate(tabla, headers=headers, tablefmt="grid"))

print(f"\nEl ciclo de vida del generador para estos parámetros es de {periodo} iteraciones.")
```
