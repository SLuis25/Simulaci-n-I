```python
# prompt: Gráfica una parabola

import matplotlib.pyplot as plt
import numpy as np

# Definir la función de la parábola
def parabola(x):
  return x**2

# Crear un rango de valores para x
x = np.linspace(-5, 5, 100)

# Calcular los valores correspondientes de y
y = parabola(x)

# Crear la gráfica
plt.plot(x, y)

# Agregar etiquetas a los ejes y título
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfica de una parábola")

# Mostrar la gráfica
plt.grid(True)
plt.show()
```
