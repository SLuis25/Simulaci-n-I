```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
```

```python
a = 1
b = 3
c = 5
M = 2/(c-a)
```

```python
def f(x):
  if a <= x <= b:
    return 2*(x - a)/((c - a)*(b - a))
  elif b < x <= c:
    return 2*(c - x)/((c - a)*(c - b))
  else:
    return 0
```

```python
def aceptacion(n):
  lista = []
  for _ in range(n):
    R1 = random()
    R2 = random()
    x = a + (c-a)*R1
    if R2 <= f(x)/M:
      lista.append(x)
  return lista
```

```python
d = aceptacion(50000)
```

```python
plt.hist(d, bins=100, density=True, alpha=0.6, label='Histograma')
x_vals = np.linspace(a, c, 500)
f_vals = [f(x) for x in x_vals]
plt.plot(x_vals, f_vals, color='red', label='Distribución Triangular teórica')
plt.title('Distribución Triangular con a=1, b=3, c=5')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```
