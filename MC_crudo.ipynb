```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sqrt(np.arctan(x))
```

```python
n = 10000
x_random = np.random.uniform(0, 1, n)  # Generar ¿n puntos aleatorios entre 0 y 1
y_values = f(x_random)  # Evaluar la función en esos puntos

# Calcular la integral aproximada
integral_approx = np.mean(y_values)

# Número de puntos en cada iteración
N_values = np.arange(1, 10001)
integral_approximations = [np.mean(f(np.random.uniform(0, 1, N))) for N in N_values]
print(f'La integral aproximada es: {integral_approx}')
```

```python
# Graficar la convergencia de la integral
plt.plot(N_values, integral_approximations, label='Aproximación de la integral')
plt.axhline(y=integral_approx, color='r', linestyle='--', label='Valor real aproximado')
plt.xlabel('Número de puntos (n)')
plt.ylabel('Valor de la integral')
plt.title('Convergencia de la integral mediante el Método de Montecarlo')
plt.legend()
plt.show()
```
