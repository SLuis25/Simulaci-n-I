```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(n):
  x=[]
  for i in range(n):
    R = rd.random()
    if R <= 0.5:
      x.append(np.sqrt(8*R)+1)
    else:
      x.append(5 - np.sqrt((1-R)*8))
  return x
```

```python
a=func(10000)
```

```python
plt.hist(a, bins=20)
```

```python
x_sorted = np.sort(a)
y_ecdf = np.arange(1, len(x_sorted)+1) / len(x_sorted)

# Grafica
plt.figure()
plt.step(x_sorted, y_ecdf, where='post')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Función de Distribución Acumulada')
plt.grid(True)
plt.show()
```
