```python
import random
import numpy as np
from numba import njit
```

```python
@njit
def dado(m):
  sumatoria = 0
  for i in range (m):
    casilla =0
    n=0
    tiro =0
    while casilla <20:
      tiro = random.randint(1,6)
      casilla = casilla + tiro
      if casilla == 3:
        casilla = 11
      if casilla == 15:
        casilla = 19
      if casilla == 13:
        casilla = 4
      if casilla == 17:
        casilla = 10
      if casilla > 20:
        casilla = 20
      n= n+1

      #print("dado: ", tiro)
      #print("casilla: ", casilla)
      #print("numero de tiros:", n)

    print("numero total de tiros:", n)
    sumatoria = sumatoria + n
  print("sumatoria: ", sumatoria)
  prom= sumatoria/m
  return("promedio de tiros para ganar: ", prom)

dado(10000)
```
