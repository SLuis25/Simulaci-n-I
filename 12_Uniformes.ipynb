```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
```

```python
# Estadísticas teóricas
mu = 2
sigma = 3
var_teorica = sigma**2
```

```python
def uniformes(n):
    lista = []
    for _ in range(n):
        # Suma de 12 U(0,1)
        suma12 = sum(random() for _ in range(12))
        # Centrar en 0 y escalar a la Normal deseada
        x = mu + sigma * (suma12 - 6)
        lista.append(x)
    return lista
```

```python
N = 100000
y = uniformes(N)
```

```python
# Estadísticas muestrales
media_emp = np.mean(y)
std_emp = np.std(y)
var_emp = np.var(y)
```

```python
print(f"Media empírica:               {media_emp:.4f}  (teórica: {mu})")
print(f"Desviación estándar empírica: {std_emp:.4f}  (teórica: {sigma})")
print(f"Varianza empírica:            {var_emp:.4f}  (teórica: {var_teorica})")
```

```python
plt.figure(figsize=(8, 5))

# Histograma normalizado a densidad
count, bins, _ = plt.hist(y, bins=50, density=True, alpha=0.6, edgecolor='black', label='Muestra (12 uniformes)')

# Puntos para la Densidad teórica
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 200) #En este rango está más del 99.99 % de la probabilidad de una Normal
F_densidad_teor = (1/(sigma * np.sqrt(2*np.pi)) * np.exp(-0.5 * ((x - mu)/sigma)**2))

plt.plot(x, F_densidad_teor, 'r-', lw=2, label='Densidad teórica')
plt.title('Simulación de $N(\\mu,\\sigma^2)$ Método 12 uniformes')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```
