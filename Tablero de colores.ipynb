# Con diccionarios

## Diccionarios en Python: Funcionamiento general

Un diccionario es una estructura de datos que permite almacenar pares de datos organizados como clave: valor (key-value pairs).

## Características principales
Clave única: Cada clave en el diccionario debe ser única. No puede haber claves repetidas.

* Valores asociados: Cada clave apunta a un valor, que puede ser de cualquier tipo (número, cadena, lista, objeto, etc.).

* Mutabilidad: Los diccionarios son mutables, es decir, pueden ser modificados después de su creación (añadir, eliminar o actualizar pares clave-valor).

* No ordenados (antes de Python 3.7): Antes de la versión 3.7, los diccionarios no garantizaban orden; a partir de Python 3.7, mantienen el orden de inserción.

* Acceso rápido: La búsqueda, inserción y eliminación de elementos se realiza en tiempo promedio cercano a $O(1)$ gracias a su implementación basada en tablas hash.

Creación:

    dic = {'clave1': valor1, 'clave2': valor2}

Acceso a valores:

    valor = dic['clave1']
Si la clave no existe, se genera un error KeyError.

Inserción o actualización:

    dic['clave3'] = valor3  # Inserta o actualiza clave3

Eliminación:

    del dic['clave1']

Verificación de existencia:

    if 'clave2' in dic:
        # hacer algo

Recorrido:
    
    for clave, valor in dic.items():
        print(clave, valor)


## Consideraciones sobre claves
Las claves deben ser objetos inmutables (por ejemplo, cadenas, números, tuplas inmutables).

No pueden ser listas o diccionarios porque estos son mutables y no hashables.

## Uso y aplicaciones
Los diccionarios son ideales para representar asociaciones, tablas de búsqueda, índices, estructuras JSON, bases de datos en memoria, entre otros.

Su eficiencia para buscar y modificar datos es una de sus mayores ventajas frente a otras estructuras como listas o tuplas.

```python
import numpy as np
```

```python
n = 2       # número de monedas
p = 0.5     # probabilidad de cara
steps = 100000  # número de lanzamientos
```

```python
caras = np.random.binomial(n, p, steps)

sectores = np.cumsum(caras) % 4
```

```python
# Inicializar diccionario con conteo en 0 para cada sector
frecuencias = {0: 0, 1: 0, 2: 0, 3: 0}

for sector in sectores:
    frecuencias[sector] += 1
```

```python
# Convertir conteos a probabilidades
for sector in frecuencias:
    frecuencias[sector] /= steps
```

```python
# Mostrar resultados
for sector, prob in frecuencias.items():
    print(f"Sector {sector}: probabilidad aproximada = {prob:.4f}")
```

Repitiendo el experimento $k$ veces

```python
import numpy as np
import matplotlib.pyplot as plt
from numba import njit  # Aceleración con Numba
```

```python
# Función acelerada con Numba para simular sectores
@njit
def simular_sectores(n, p, steps):
    resultados = np.random.binomial(n, p, steps)
    posiciones = np.empty(steps, dtype=np.int32)
    suma = 0
    for i in range(steps):
        suma += resultados[i]
        posiciones[i] = suma % 4
    return posiciones
```

```python
# Ejecutar múltiples simulaciones para calcular promedios por sector a lo largo del tiempo
def simulacion_masiva(n, p, steps, k):
    acumuladas = np.zeros((4, steps))

    for _ in range(k):
        sectores = simular_sectores(n, p, steps)
        conteos = np.zeros(4)
        acumulado = np.zeros(steps)

        for i in range(steps):
            conteos[sectores[i]] += 1
            for j in range(4):
                acumuladas[j, i] += conteos[j] / (i + 1)

    # Promedio de todas las simulaciones
    return acumuladas / k
```

```python
# Parámetros
n = 2
p = 0.5
steps = 1000  # Usamos 1000 pasos por eficiencia gráfica
k = 10000     # Repeticiones del experimento
```

```python
# Ejecutar simulación masiva
evoluciones_promedio = simulacion_masiva(n, p, steps, k)
```

```python
# Colores personalizados
custom_colors = ['red', 'orange', 'green', 'blue']
```

```python
# Graficar evolución promedio para todos los sectores
for sector in range(4):
    plt.plot(evoluciones_promedio[sector], label=f'Sector {sector}', color=custom_colors[sector])

plt.axhline(y=0.25, color='black', linestyle='--', label='Valor esperado (0.25)')
plt.xlabel('Número de pasos')
plt.ylabel('Probabilidad acumulada promedio')
plt.title(f'Promedio de {k} simulaciones para cada sector')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Imprimir probabilidades finales aproximadas por sector
for sector in range(4):
    final_prob = evoluciones_promedio[sector][-1]
    print(f"Probabilidad final promedio - Sector {sector}: {final_prob:.4f}")

```
